```python
from basic_kmeans import KMeansClustering
import matplotlib.pyplot as plt

import numpy as np
```

# EXEMPLO DO K-Means


##  Array inicial:
### [[1, 2], [1, 4], [1, 0], [4, 2], [4, 4], [4, 0]]



##    Inicialização dos Centróides:
###   O algoritmo começa inicializando aleatoriamente os centróides dentro do intervalo dos dados fornecidos. 
###   Suponha que os centróides iniciais sejam [1, 2] e [4, 4].


##     Atribuição de Clusters:
###    Cada ponto de dados é atribuído ao centróide mais próximo. 
###    [1, 2], [1, 4], e [1, 0] são atribuídos ao centróide [1, 2].
###    [4, 2], [4, 4], e [4, 0] são atribuídos ao centróide [4, 4].

##     Recalcular Centróides:
###    Os novos centróides são calculados como a média dos pontos atribuídos a cada cluster. 
###    O algoritmo continua iterando até que os centróides não mudem significativamente entre as iterações. 

##     Resultado Final:
###    Os pontos de dados são atribuídos aos clusters finais com base nos centróides finais. Suponde que os centróides finais são [1, 2] e [4, 2].
###    Os pontos [1, 2], [1, 4], e [1, 0] pertencem ao cluster com centróide [1, 2].
###    Os pontos [4, 2], [4, 4], e [4, 0] pertencem ao cluster com centróide [4, 2].









```python
# Generate random data points
random_points = np.random.randint(0, 500, size=(200, 2))
# Create an instance of KmeansClustering and fit the model to the data
kmeans = KMeansClustering()
initial_centroids, intermediate_centroids, final_centroids, intermediate_labels, final_labels = kmeans.fit(random_points)



```

```python
# previsualização dos dados
plt.scatter(random_points[:, 0], random_points[:, 1], c='gray')
plt.show()
```

```python

plt.figure(figsize=(6, 6))
plt.scatter(random_points[:, 0], random_points[:, 1], c='gray')
plt.scatter(initial_centroids[:, 0], initial_centroids[:, 1], c='red', marker='*', s=250)
plt.title('Centroides iniciais')
plt.show()
```

```python
if len(intermediate_centroids) > 0:
    plt.figure(figsize=(6, 6))
    plt.scatter(random_points[:, 0], random_points[:, 1], c=intermediate_labels[0], cmap='viridis')
    plt.scatter(intermediate_centroids[0][:, 0], intermediate_centroids[0][:, 1], c='red', marker='*', s=250)
    plt.title('Cluster Inicial')
    plt.show()
```

```python
# Final centroids
plt.figure(figsize=(6, 6))
scatter = plt.scatter(random_points[:, 0], random_points[:, 1], c=final_labels, cmap='viridis')
plt.scatter(final_centroids[:, 0], final_centroids[:, 1], c='red', marker='*', s=250)
plt.title('Cluster final')

# Add legend
legend1 = plt.legend(*scatter.legend_elements(), title="Clusters")
plt.gca().add_artist(legend1)

plt.show()
```
